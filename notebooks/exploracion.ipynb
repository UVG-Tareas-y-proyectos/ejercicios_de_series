# Ejercicio 1 - Temperatura Guatemala

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv("../data/raw/guatemala_temperatura.csv", parse_dates=["month"])
df.head()
```

```python
df.describe()
```

```python
capas = [
    "temperature_2m_c",
    "soil_temperature_layer_1_c",
    "soil_temperature_layer_2_c",
    "soil_temperature_layer_3_c",
    "soil_temperature_layer_4_c",
]

# extremos de cada capa (2m = superficie)
for c in capas:
    fmax = df.loc[df[c].idxmax()]
    fmin = df.loc[df[c].idxmin()]
    print(f"{c}: max={fmax[c]:.2f} ({fmax['month'].date()})  min={fmin[c]:.2f} ({fmin['month'].date()})")
```

```python
# tendencia: regresion lineal contra el tiempo en años
df["t"] = (df["year"] - df["year"].min()) + (df["month_num"] - 1) / 12

for c in capas:
    slope, intercept, r, p, se = stats.linregress(df["t"], df[c])
    print(f"{c}: {slope:.4f} C/año  (p={p:.4g})")
```

```python
df["temp_2m_rolling"] = df["temperature_2m_c"].rolling(12).mean()

plt.figure(figsize=(10, 5))
plt.plot(df["month"], df["temperature_2m_c"], alpha=0.5, label="temp 2m")
plt.plot(df["month"], df["temp_2m_rolling"], color="red", label="promedio movil 12m")
plt.legend()
plt.title("Temperatura promedio a 2m - Guatemala")
plt.savefig("../data/processed/tendencia_temp2m.png")
plt.show()
```
